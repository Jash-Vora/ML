**Imbalanced dfset:**

A dfset with an unequal class distribution

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dfset to pandas dfFrame
credit_card_df = pd.read_csv('../Datasets/credit_data.csv')
```

```python
# first 5 rows of the dfframe
credit_card_df.head()
```

```python
credit_card_df.tail()
```

```python
# distribution of the two classes
credit_card_df['Class'].value_counts()
```

This is Highly Imbalanced dfset

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# separating the legit and fraudulent transactions
legit = credit_card_df[credit_card_df.Class == 0]
fraud = credit_card_df[credit_card_df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Build a sample dfset containing similar distribution of Legit & Fraudulent Transactionds

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the Two dfFrames

```python
new_dfset = pd.concat([legit_sample, fraud], axis = 0)
```
