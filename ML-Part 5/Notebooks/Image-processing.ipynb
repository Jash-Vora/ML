Converting rgb images to grayscale using opencv

```python
import cv2
```

```python
img = cv2.imread('../Datasets/painting-mountain-lake-with-mountain-background_188544-9126.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```

```python
cv2.imshow('Grayscale Image', grayscale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('img_grayscale.jpg',grayscale_img)
```
