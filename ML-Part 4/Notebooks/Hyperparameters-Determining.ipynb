```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
```

```python
breast_dataset = load_breast_cancer()
```

```python
print(breast_dataset)
```

```python
breast_df = pd.DataFrame(breast_dataset.data,columns=breast_dataset['feature_names'])
```

```python
print(breast_df)
```

```python
breast_df['target'] = breast_dataset.target
```

```python
breast_df.head()
```

```python
breast_df.tail()
```

```python
breast_df.target.value_counts()
```

1 --> Benign

0 --> Malignant

```python
X = breast_df.drop(columns='target',axis=1)

Y = breast_df.target
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

Grid Search CV

```python
model = SVC()
```

```python
# hyperparameters

parameters = {
    'kernel': ['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# grid search
classifier = GridSearchCV(model,parameters,cv=5)
```

```python
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
best_params = classifier.best_params_

print(best_params)
```

```python
highest_acc = classifier.best_score_

print(highest_acc)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
print(grid_search_result)
```

RandomizedSearchCV

```python
model = SVC()
```

```python
parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
classifier = RandomizedSearchCV(model,parameters,cv=5)
```

```python
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
best_params = classifier.best_params_
print(best_params)
```

```python
highest_acc = classifier.best_score_

print(highest_acc)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
print(randomized_search_result)
```
