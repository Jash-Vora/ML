```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,-3,-6,scale_units='xy',angles='xy',scale = 1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,3,-2])

v3 = v1 + v2
print(v3)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,3,-1,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two vectors

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,3,-2])

v3 = v1 - v2
print(v3)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,3,-1,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by scalar

```python
v1 = np.asarray([0,0,2,3])
v2 = 2*v1
print(v2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
v1 = np.asarray([0,0,2,3])
v2 = -0.5*v1
print(v2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,-1,-1.5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
